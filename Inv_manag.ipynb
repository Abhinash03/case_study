```python
# Group by 'Invoice' and sum 'Total_Sales' to get total revenue per invoice
invoice_totals = df.groupby('Invoice')['Total_Sales'].sum().reset_index()
```

```python
#"Which 10 invoices have the highest total sales amounts?"
top10_invoices = invoice_totals.sort_values(by='Total_Sales', ascending=False).head(10)
print(top10_invoices)
```

```python
# Which weekday generates the highest sales revenue?
weekday_sales = df.groupby('Weekday')['Total_Sales'].sum().sort_values(ascending=False)
print(weekday_sales)
```

```python
#In which month did the company generate the most revenue?
monthly_revenue = df.groupby('Month')['Total_Sales'].sum().sort_values(ascending=False)
print(monthly_revenue)
```

```python
#What is the average purchase amount per transaction (invoice)?
avg_transaction = df.groupby('Invoice')['Total_Sales'].sum().mean()
print(f"Average Sales per Transaction: ₹{avg_transaction:.2f}")
```

```python
#Which invoices have unusually high sales values (potential fraud or big clients)?
high_value_invoices = df.groupby('Invoice')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print(high_value_invoices)
```

```python
#What is the return rate (Invoices starting with 'C')?
total_invoices = df['Invoice'].nunique()
return_invoices = df[df['Invoice'].str.startswith('C')]['Invoice'].nunique()
return_rate = (return_invoices / total_invoices) * 100
print(f"Return Rate: {return_rate:.2f}%")
```

```python
#Visualization-Based 
#What are the top 10 best-selling products by total revenue?
import matplotlib.pyplot as plt
import seaborn as sns

top_products = df.groupby('Description')['Total_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
plt.title('Top 10 Products by Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Product Description')
plt.tight_layout()
plt.show()
```

```python
#Which day of the week generates the most sales?
weekday_sales = df.groupby('Weekday')['Total_Sales'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

plt.figure(figsize=(8, 5))
sns.barplot(x=weekday_sales.index, y=weekday_sales.values, palette='coolwarm')
plt.title('Sales by Day of the Week')
plt.ylabel('Total Sales')
plt.xlabel('Weekday')
plt.show()
```

```python
#How do monthly sales trend over the year?
monthly_sales = df.groupby('Month')['Total_Sales'].sum()

plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='magma')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("online_retail.csv")  # Replace with your actual file path



# Select only numerical columns for correlation
numerical_cols = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
correlation_matrix = numerical_cols.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Step 1: Keep only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Step 2: Drop or fill NaNs/Infs
numeric_df = numeric_df.replace([np.inf, -np.inf], np.nan)  # Convert infs to NaN
numeric_df = numeric_df.dropna()  # Drop rows with NaNs

# Step 3: Add constant
X = add_constant(numeric_df)

# Step 4: Calculate VIF
vif = pd.DataFrame()
vif['Feature'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Step 5: Show VIF
print(vif)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2, SelectKBest
import numpy as np

# Step 1: Select original categorical columns
cat_cols = ['Invoice', 'StockCode', 'Description']
cat_df = df[cat_cols].copy()

# Step 2: Encode and store LabelEncoders
encoded_cat_df = pd.DataFrame()
label_encoders = {}

for col in cat_df.columns:
    le = LabelEncoder()
    encoded_cat_df[col] = le.fit_transform(cat_df[col].astype(str))
    label_encoders[col] = le  # Save encoder

# Step 3: Create a dummy 'Return' column (0 or 1)
# This is just for testing – replace with real return data later
df['Return'] = np.random.choice([0, 1], size=len(df))
y = df['Return']

# Step 4: Apply Chi-Square test
chi_selector = SelectKBest(score_func=chi2, k='all')
chi_selector.fit(encoded_cat_df, y)

# Step 5: Create and display Chi-Square results
chi_scores = pd.DataFrame({
    'Feature': encoded_cat_df.columns,
    'Chi2 Score': chi_selector.scores_,
    'P-Value': chi_selector.pvalues_
})

# Output
print("Encoded Categorical Features:")
print(encoded_cat_df.head())

print("\nChi-Square Test Results:")
print(chi_scores)
```

🧠 Goal of This Code:
Encode categorical features so that they can be used in statistical models.

Apply Chi-Square test to check the importance of those features in predicting Return.
